```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
data1 =pd.read_csv("california_cities.csv")

latd, longd = data1['latd'], data1['longd']
population, area= data1['population_total'], data1['area_total_km2']
data1.head()
```

```python
x= longd
y = latd
plt.scatter(x,y, s=area, c= np.log10(population), cmap='PiYG_r')
plt.title('California Cities: Area and Population')
plt.xlabel('Lognitude')
plt.ylabel('Latitude')
plt.colorbar(label='Log$_{10}$(population)')
plt.show()
```
